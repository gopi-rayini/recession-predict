```python
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
# Which ETF performs the best in an economic recession?
# What is the expected rate of growth during the 6 months of recessions for each ETF?

# Import data from CSV for MSFT
spydf = pd.read_csv('data_files/SPY Daily Historical 1993.csv')
xlvdf = pd.read_csv('data_files/XLV Daily Historical 1999.csv')
iykdf = pd.read_csv('data_files/IYK Daily Historical 2001.csv')
ratesdf = pd.read_csv('data_files/DFF.csv')

# Gain a feel for the data
# display(msftdf.head)
# print(df.tail)
# print(df.shape)
# print(fedratedf.info())

# Look at columns and reduce to ones we care about
# display(fdf.columns)
```

```python
spydf["SPY Price Change"] = ((spydf["Close"] - spydf["Open"])/spydf["Open"])*100
xlvdf["XLV Price Change"] = ((xlvdf["Close"] - xlvdf["Open"])/xlvdf["Open"])*100
iykdf["IYK Price Change"] = ((iykdf["Close"] - iykdf["Open"])/iykdf["Open"])*100
```

```python
spydf.drop(["Open","Close","High","Low","Adj Close"], axis=1, inplace=True)
xlvdf.drop(["Open","Close","High","Low","Adj Close"], axis=1, inplace=True)
iykdf.drop(["Open","Close","High","Low","Adj Close"], axis=1, inplace=True)
```

```python
df = spydf.merge(right=xlvdf, on='Date', how='inner')
df = df.merge(right=iykdf, on='Date', how='inner')
ratesdf.rename(columns={"DATE":"Date"}, inplace=True)
ratesdf.info()
df = df.merge(right=ratesdf, on='Date', how='inner')
df.rename(columns={"Volume_x":"SPY Volume", "Volume_y":"XLV Volume", "Volume":"IYK Volume"}, inplace=True)
df.describe()
```

```python
df.to_csv('data_files/data.csv', index=False)
```
